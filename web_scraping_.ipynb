```python
from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  counted = (series['index']
                .value_counts()
              .reset_index(name='counts')
              .rename({'index': 'index'}, axis=1)
              .sort_values('index', ascending=True))
  xs = counted['index']
  ys = counted['counts']
  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = _df_1.sort_values('index', ascending=True)
_plot_series(df_sorted, '')
sns.despine(fig=fig, ax=ax)
plt.xlabel('index')
_ = plt.ylabel('count()')
```

---

















**key insight No 1 :Total number of Parties.**

---



```python
total_constituencies = df["Party"].nunique()
print("key insight No 1 : Total number of Parties in Lok Sabha election .")
print(" Total number of Parties:", total_constituencies)
```

**key insight No 2: Number of parties with names starting with 'A'**

---



```python
num_parties_starting_with_a = df[df["Party"].str.startswith("A")]["Party"].nunique()
print("Number of parties with names starting with 'A':", num_parties_starting_with_a)
print("Parties starting with 'A':")
for party in list(df[df["Party"].str.startswith("A")]["Party"].unique()):
    print(party)
```

**key insight No 3: Party with the most seats won**

---



```python
most_seats_party = df.loc[df["Won"].idxmax(), "Party"]
print("Party with the most seats won:", most_seats_party)
```

**key insight No 4 :Total number of seats for each party**

---



```python
df.loc[:, 'Total'] = pd.to_numeric(df['Total'], errors='coerce')
total_seats_each_party = df.groupby("Party")["Total"].sum()
print("Insight 15: Total number of seats for each party:\n", total_seats_each_party)
```

**key insight No 5: Party with the most seats won and leading combined**

---



```python
df.loc[:, "Won"] = pd.to_numeric(df["Won"])
df.loc[:, "Leading"] = pd.to_numeric(df["Leading"])
df.loc[:, "Won_Leading_Combined"] = df["Won"] + df["Leading"]
most_seats_won_leading_combined_party = df.loc[df["Won_Leading_Combined"].idxmax(), "Party"]
print(" Party with the most seats won and leading combined:", most_seats_won_leading_combined_party)
```

**key insight No 6 :Party with the lowest seats won**

---


```python
lowest_seats_won_party = df.loc[df["Won"].idxmin(), "Party"]
print(" Party with the lowest seats won:", lowest_seats_won_party)
```

**key insight No 7:Average seats won per party**

---



```python
average_seats_won = df["Won"].mean()
print(" Average seats won per party:", average_seats_won)
```

**key insight No 8 : Total seats won by all parties**

---





```python
total_seats_won = df["Won"].sum()
print(" Total seats won by all parties:", total_seats_won)
```

**key insight No 9: Total seats leading by all parties**

---



```python
total_seats_leading = df["Leading"].sum()
print(" Total seats leading by all parties:", total_seats_leading)
```

**key insight No 10 : Party with the highest average seats won per contest**

---



```python
highest_average_won_party = df.groupby("Party")["Won"].mean().idxmax()
print(" Party with the highest average seats won per contest:", highest_average_won_party)
```
